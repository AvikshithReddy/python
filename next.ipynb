```python
he = df.groupby(['job_country', 'job_title_short'])['salary_year_avg'].agg('median').dropna()
```

```python
top_countries = df['job_country'].value_counts().head(6).index
he = he.loc[top_countries]

job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist'] # 'Senior Data Analyst', 'Senior Data Engineer', 'Senior Data Scientist']
he = he[job_titles]

he.plot(kind='bar')
```

```python
job_country_salary = df.pivot_table(
    values='salary_year_avg',        # The data to aggregate
    index='job_country',             # Rows (index of pivot table)
    columns='job_title_short',       # Columns (pivot table headers)
    aggfunc='median'                 # Aggregation function
)

job_country_salary
```

```python
top_countries = df['job_country'].value_counts().head(6).index
job_country_salary = job_country_salary.loc[top_countries]

job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist'] # 'Senior Data Analyst', 'Senior Data Engineer', 'Senior Data Scientist']
job_country_salary = job_country_salary[job_titles]
```

```python
job_country_salary.plot(kind='bar')
plt.ylabel('Median Salary ($USD)')
plt.xlabel('')
plt.title('Median Salary by Country and Job Title')  
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
df_skills = df.explode('job_skills')
df_skills
```

```python
skills_count = df_skills.groupby(['job_skills', 'job_title_short']).size()
skills_count
```

```python
df_skills_count = skills_count.reset_index(name='skill_count') # we need to rename the main Series value to count

df_skills_count
```

```python
df_skills_count.sort_values(by='skill_count',  inplace=True, ascending=False)

df_skills_count
```

```python
job_title = 'Data Engineer'
head_value = 10

df_skill_final = df_skills_count[df_skills_count['job_title_short'] == job_title].head(head_value)

df_skill_final
```

```python
df_skill_final.plot(kind='bar', x='job_skills', y='skill_count')
```
