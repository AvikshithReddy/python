## Skills trends

```python
import ast
```

```python
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Now explode
```

```python
df_DA = df[df['job_title_short'] == 'Data Scientist'].copy()
df_DA
```

```python
df_DA_explode= df_DA.explode('job_skills')
df_DA_explode
```

```python
df_DA_pivot = df_DA_explode.pivot_table(index='job_posted_month', columns='job_skills',  aggfunc='size', fill_value=0)

df_DA_pivot
```

```python
# Create a total row
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()
# Order columns by total count
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]
# Drop total row
df_DA_pivot = df_DA_pivot.drop('Total')

df_DA_pivot
```

```python
df_DA_pivot.iloc[:, :7].plot(kind='line')

plt.title('Top 7 Skills for Data Analysts per Month')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```

```python
df_DA_pivot.iloc[:, :5].plot(
    kind='line',
    linewidth=4,
    linestyle=':', # options: '-', '--', '-.', ':'
    colormap='viridis', # options: 'viridis', 'plasma', 'inferno', 'magma', 'cividis' 
    markersize=5,
    marker='o', # options: 'o', 'x', '+', '*', 's', ',', '.', '1', '2', '3', '4', ...
    figsize=(10, 5)
)

plt.title('Top 5 Skills for Data Analysts per Month')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```
